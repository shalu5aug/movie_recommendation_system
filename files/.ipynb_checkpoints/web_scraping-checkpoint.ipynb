```python
class get_movie_data:
    
    def __init__(self,url):
        self.url = url
    
    def web_scrape(self):
        df = pd.DataFrame()
        webpage = requests.get(self.url).text
        soup = BeautifulSoup(webpage, 'lxml')
        table = soup.find_all('table', class_='wikitable sortable')
        for i in range(len(table)):
            df1 = pd.read_html(str(table[i]))[0]
            df = df.append(df1,ignore_index=True)
        return(df)
        
```

```python

movie_df = pd.DataFrame()
#get data from 2018 to 2022
for i in range(2017,2023):
    url = f'https://en.wikipedia.org/wiki/List_of_American_films_of_{i}'
    data=get_movie_data(url)
    fin_df=data.web_scrape()
    
    movie_df = movie_df.append(fin_df,ignore_index=True)
    
```

```python
from tmdbv3api import TMDb
from tmdbv3api import Movie
```

```python
tmdb = TMDb()
tmdb.api_key = config.MOVIE_API_KEY
tmdb.language = 'en'
```

```python
genres = []
```

```python
gnrs_url = f'https://api.themoviedb.org/3/genre/movie/list?api_key={tmdb.api_key}&language=en-US'
response = requests.get(gnrs_url).json()
genres = response['genres']
```

```python
temp_df = pd.DataFrame(columns=['genres','id','Title'])
temp_df['Title'] = movie_df['Title']

tmdb_movie = Movie()
tmdb_movie.language='en'
for ind  in range(len(movie_df)):
    results=tmdb_movie.search(term=temp_df.loc[ind]['Title'])
    if results:

        if results[0]['genre_ids']:
            temp_df.loc[ind,'genres'] = results[0]['genre_ids']
        if results[0]['id']:
            temp_df.loc[ind,'id'] = results[0]['id']
```

```python
temp_df[temp_df['Title']=='Wildcat']
```

```python
movie_df = pd.merge(movie_df, temp_df, on='Title')
```

```python
movie_df.dropna(inplace=True)
```

```python
movie_df = movie_df.reset_index(drop=True)
```

```python
gen = {}
for item in genres:
    gen[item['id']] = item['name']
```

```python
# gen_lst = movie_df['genres'].apply(lambda item_genres :{itm :value for (itm,value) in gen.items() if itm in item_genres})
```

```python
movie_df['genres'] = movie_df['genres'].apply(lambda item :[gen[i] for i in  item])
```

```python
movie_df['genres'] = movie_df['genres'].apply(lambda genres: " ".join(genres))
```
