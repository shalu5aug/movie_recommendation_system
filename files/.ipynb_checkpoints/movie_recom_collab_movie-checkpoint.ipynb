```python
# ratings = pd.read_csv('data/movie_lens/ratings.csv')
ratings = pd.read_table('movie_tutorial/data/ml-1m/ratings.dat',
                        sep='::', header=None, 
                        names=['user_id', 'movie_id', 'rating', 'timestamp'])
```

```python
# lets start with movieid = 1(toy story)
```

```python
flt = (( ratings['movie_id']==1) & (ratings['rating']>4))
```

```python
sim_users = ratings.loc[flt]['user_id'].unique()
```

```python
sim_users
```

```python
#get all the movies which similar users liked
```

```python
flt = ((ratings['user_id'].isin(sim_users)) & (ratings['rating']>4))
```

```python
sim_user_movies = ratings.loc[flt]['movie_id'].unique()
```

```python
sim_user_movies
```

```python
sim_user_movies_matrix = ratings.loc[flt]['movie_id'].value_counts()
```

```python
sim_user_movies_matrix = pd.DataFrame(sim_user_movies_matrix).reset_index().rename(columns={'index':'movie_id','movie_id':'tot_users'})
```

```python
sim_user_movies_matrix['tot_users'] = sim_user_movies_matrix['tot_users']/len(sim_users)
```

```python
#get those movie which 70% of similar users liked
```

```python
flt = sim_user_movies_matrix['tot_users']>.3
sim_user_movies_matrix  = sim_user_movies_matrix.loc[flt]
sim_user_movies_matrix
```

```python
#get all the movies from similar users movies which other users liked 
```

```python
 flt = (ratings['movie_id'].isin(sim_user_movies_matrix['movie_id']) & (ratings['rating']>4))
```

```python
other_users_movies_matrix =  ratings.loc[flt]
other_users_movies_matrix = pd.DataFrame(other_users_movies_matrix['movie_id'].value_counts()/len(other_users_movies_matrix['user_id'].unique())).reset_index().rename(columns={'index':'movie_id','movie_id':'tot_users'})
```

```python
other_users_movies_matrix
```

```python
final_matrix = pd.merge(sim_user_movies_matrix,other_users_movies_matrix, on='movie_id', how='left')
```

```python
final_matrix.dropna()
```

```python
final_matrix['perc'] = final_matrix['tot_users_x']/final_matrix['tot_users_y']
```

```python
final_matrix.sort_values(by='perc', ascending=False)
```
