```python
lst_movie_3 = [movieId for movieId in user_3['movieId']]
```

```python
lst_movie_3 = list(set(lst_movie_3))
```

```python
flt =(ratings['movieId'].isin(lst_movie_3) & (ratings['rating']>=4) & (ratings['userId']!=3))
```

```python
sim_user = list(set(sim_user))
```

```python
len(sim_user)
```

```python
# find recommended movies by similar users with rating >=3.5
```

```python
flt = ((ratings['userId'].isin(sim_user)) & (ratings['rating']>=5) & (ratings['movieId'].isin(lst_movie_3)==False))
```

```python
recom_movies = ratings.loc[flt]
```

```python
recom_movies = recom_movies['movieId'].value_counts()
```

```python
recom_movies = pd.DataFrame(recom_movies).reset_index()
```

```python
recom_movies = recom_movies.rename(columns={'index':'movieId','movieId':'tot_users'})
```
